```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate

# temperature -> 모델의 창의력 수치 0.1 은 별로 창의적이지 않은 상태 0.9는 매우 창의적인 상태
chat = ChatOpenAI(temperature=0.1)
```

```python
from langchain.schema import BaseOutputParser


class CommaOutputParser(BaseOutputParser):
    def parse(self, text):
        items = text.split(",")
        return list(map(str.strip, items))
    
p = CommaOutputParser()

p.parse("Hello, How, Are, You")
```

```python
template = ChatPromptTemplate.from_messages([
    ("system", "You are a list generathing machine. Everything you are asked will be anwered with a comma seperated list {max_items} in lower case. Do NOT reply with anything else."), 
    ("user", "{question}")
])
```

```python
# lang chain ...
chain = template | chat | CommaOutputParser() 
chain.invoke({
    "max_items" : 5,
    "question" : "What are the colors?"
})
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(temperature=0.4, streaming=True, callbacks=[
                  StreamingStdOutCallbackHandler()])
chef_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a world-class international chef. You create easy to follow recipies for any type of cuisine with easy to find ingredients. reply language is only korea word"),
    ("user", "I want to cook {cuisine} food.")
])

veg_chef_prompt = ChatPromptTemplate.from_messages(
    [("system", "You are a vegetarian chef specialized on making traditional recipes vegetarian. You find alternative ingredients and explain their preparation. You don't radically modify the recipe. If there is no alternative for a food just say you don't know how to recipe it."), ("human", "{recipe}")])

chef_chain = chef_prompt | chat

veg_chain = veg_chef_prompt | chat

final_chain = {"recipe": chef_chain} | veg_chain

final_chain.invoke({
    "cuisine": "korean"
}).content
```
